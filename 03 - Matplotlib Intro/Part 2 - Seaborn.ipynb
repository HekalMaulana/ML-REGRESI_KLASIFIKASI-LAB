# Seaborn & Pandas Dataframe

## Statistical Summary

```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

## Missing Value

```python
plt.figure(figsize=(6, 6))
sns.heatmap(data=df.isna(), cmap="Blues", cbar=False);
```

```python
plt.figure(figsize=(6, 6))
sns.heatmap(data=~df.isna(), cmap="Blues", cbar=False);
```

## Correlation Heatmap

```python
plt.figure(figsize=(6, 6))
sns.heatmap(data=df.corr(numeric_only=True), cmap="bwr", cbar=False, vmin= -1, vmax= 1, annot=True, square=True, fmt=".1f");
```

## Pairplot

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue="shelf", vars=["calories", "fat", "rating"])
```

```python
sns.pairplot(df, hue="shelf", vars=["calories", "fat", "rating"], kind="reg")
```

# Categorical Data

## Persebaran

```python
sns.catplot(x="mfr", y="rating", data=df, hue="mfr");
```

```python
sns.boxplot(x="mfr", y="rating", data=df, hue="mfr");
```

```python
sns.violinplot(x="mfr", y="rating", data=df, hue="mfr");
```

## Count

```python
sns.catplot(x="mfr", data=df, kind="count")
```

```python
sns.countplot(x="mfr", data=df, hue="shelf")
```
