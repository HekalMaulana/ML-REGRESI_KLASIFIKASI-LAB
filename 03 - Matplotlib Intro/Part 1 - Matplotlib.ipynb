```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisialiasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes
plt.plot(X, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes
plt.plot(X, y1)
plt.plot(X, y2)
```

# Scatter Plot

```python
df = pd.read_csv("data/circle_data.csv", sep=";")
df.head()
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X, y)
plt.axis("equal")
```

```python
plt.scatter(X, y, c="r", s=10)
plt.axis("equal");
```

```python
plt.scatter(X, y, c=rad, s=rad*90, cmap="seismic")
plt.axis("equal");
```

# Histogram

```python
numbers = np.random.rand(10000)
plt.hist(numbers, bins=25);
```

```python
numbers = np.random.randn(100000)
plt.hist(numbers, bins=80);
```

```python
numbers = np.random.randn(100000)
plt.hist(numbers, bins=80, density=1, cumulative=True);
```

# Subplot

## MATLAB Style

```python
plt.figure(figsize=(10, 7))

plt.subplot(231)
plt.hist(numbers, bins=80, density=1, cumulative=True);

plt.subplot(233)
plt.scatter(X, y, c="r", s=10)
plt.axis("equal");

plt.subplot(235)
plt.scatter(X, y, c=rad, s=rad*90, cmap="seismic")
plt.axis("equal");

plt.subplot(236)
plt.hist(numbers, bins=80, density=1, cumulative=True);
```

## MATPLOTLIB Style

```python
fig, ax = plt.subplots(2, 3, figsize=(10, 7))

ax[0, 0].hist(numbers, bins=80, density=1, cumulative=True)

ax[0, 1].axis("off")

ax[0, 2].scatter(X, y, c=rad, s=rad*90, cmap="seismic")
ax[0, 2].axis("equal")

ax[1, 0].axis("off")

ax[1, 1].hist(numbers, bins=80, density=1, cumulative=True)

ax[1, 2].scatter(X, y, c="r", s=10)
ax[1, 2].axis("equal");
```

```python
fig, ax = plt.subplots(2, 3, figsize=(10, 7), sharex=True)

ax[0, 0].hist(numbers, bins=80, density=1, cumulative=True)

ax[0, 1].axis("off")

ax[0, 2].scatter(X, y, c=rad, s=rad*90, cmap="seismic")
ax[0, 2].axis("equal")

ax[1, 0].axis("off")

ax[1, 1].hist(numbers, bins=80, density=1, cumulative=True)

ax[1, 2].scatter(X, y, c="r", s=10)
ax[1, 2].axis("equal");
```
